```python
import pandas as pd
```

```python
def make_latex_lines(a):
  text = f"{a[0]}"

  for i in range(1, len(a)):
    if (isinstance(a[i], str)):
      text += f" & {a[i]}"
    else:
      text += f" & {round(a[i], 4)}"

  text += r" \\" + "\n"
  return text

def make_latex_tab(df, table_title = ""):
  table_len = "l"*len(df.columns)
  start = r"""
\begin{table*}[!htb]
\centering""" + r"""
\caption{""" + f"{table_title}" + r"}" + r"""
\label{tab:table_name}
\begin{tabular*}{\textwidth}{@{\extracolsep{\fill}}""" + f"{table_len}" + "}"

  text = r"""
\hline
"""

  columns = ""
  for i in range(1, len(df.columns)):
    columns += f"""& {df.columns[i]}  """
  
  columns += r""" \\ 
\hline
"""

  text += columns

  for i in range(1, len(df) + 1):
    text += make_latex_lines(df[i-1:i].to_numpy()[0])
  
  end = r"""\hline
\end{tabular*}
\fonte{Autor.}
\end{table*}
"""

  return start + text + end
```

```python
sheet_names = ["Param_1", "Param_2", "Param_3", "Eq 2", "Eq 2 Tabela", "Eq 3", "Eq 3 Tabela", "Eq 6", "Eq 7", "Eq 20", "Eq 21", "Eq 22", "Eq 23", "Eq 29"]

index = 12
print(f"A sheet escolhida foi: {sheet_names[index]}")

df = pd.read_excel("Params.xlsx", sheet_name = sheet_names[index])

print(make_latex_tab(df))
```
